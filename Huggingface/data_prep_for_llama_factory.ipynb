```python
import json

def convert_jsonl_to_new_format(input_file, output_file):
    # 打开输入文件
    with open(input_file, 'r') as file:
        lines = file.readlines()

    # 准备新格式的数据
    new_data = []
    for line in lines:
        # 解析每一行 JSONL
        data = json.loads(line)

        real_input = ""
        try:
            real_input = real_input + data["oui_friendly"]
        except:
            pass
        
        try:
            real_input = real_input + data["dhcp_hostname"]
        except:
            pass
        
        try:
            real_input = real_input + data["domains"]
        except:
            pass
        
        try:
            real_input = real_input + data["netdisco_info"]
        except:
            pass
        
        try:
            real_input = real_input + data["name"]
        except:
            pass
        
        try:
            real_input = real_input + data["model_name"]
        except:
            pass
        
        # 构建新格式的记录
        new_record = {
            "instruction": "you are a classifier, read the contents and tell me the approximate vendor of the IoT device with the following information",
            "input": real_input,
            "output": data["output"]
        }

        # 添加到新数据列表中
        new_data.append(new_record)

    # 将新数据写入输出文件
    with open(output_file, 'w') as file:
        json.dump(new_data, file, indent=4)

# 调用函数
input_file = 'cs_eval.jsonl'  # 替换为您的输入文件名
output_file = 'cs_eval_new.json'  # 替换为您想要的输出文件名
convert_jsonl_to_new_format(input_file, output_file)
```
