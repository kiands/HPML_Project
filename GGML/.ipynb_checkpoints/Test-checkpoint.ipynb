```python
# important: load packages
from langchain.callbacks.manager import CallbackManager
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
from langchain.chains import LLMChain
from langchain.llms import LlamaCpp
from langchain.prompts import PromptTemplate
```

```python
# callback_manager = CallbackManager([StreamingStdOutCallbackHandler()])
```

```python
n_gpu_layers = -1  # Metal set to 1 is enough. -1 means all layers loaded to VRAM
n_batch = 512  # Should be between 1 and n_ctx, consider the amount of RAM of your Apple Silicon Chip.
# Make sure the model path is correct for your system!
llm = LlamaCpp(
    model_path="ggml-model-7b-q4_0.gguf",
    n_gpu_layers=n_gpu_layers,
    n_batch=n_batch,
    n_ctx=2048,
    # temperature=0.2,
    f16_kv=True,  # MUST set to True, otherwise you will run into problem after a couple of calls
    temperature = 0, # [0,1], float. Higher temperature, more fixed output (less creativity)
    # callback_manager=callback_manager,
    verbose=False,  # Verbose is required to pass to the callback manager,
)
```

single test: 'dhcp_hostname'

```python
columns_to_copy = ['device_id', 'dhcp_hostname']
df_llama2_dhcp = loaded_df[loaded_df['dhcp_hostname'].notna()][columns_to_copy].copy()
```

```python
df_llama2_dhcp.head(5)
```

```python
# Open the file in read mode ('r')
with open('chat-with-bob-vendor-prediction-dhcp.txt', 'r') as file:
    # Read the entire content of the file into a variable
    template_dhcp = file.read()

# Now, file_content contains the contents of the file
# print(template)
```

```python
prompt_dhcp = template_dhcp.rstrip() + ' If a user typed \'yeelink-light-color2_miio87609423\', which is a technology vendor or a technology product name. What is the vendor the user typed? Consider that the user may misspell the vendor or provide invalid input. \nBob: '
print(prompt_dhcp)
```

```python
# bad prompt: verbose output
llm(prompt_dhcp)
```

```python
# The code used to see how much time it needs to process a few contents
import time

start_time = time.time()

# Counter for printing messages
row_counter = 0

# Apply the function to each row
for index, row in df_llama2_dhcp.iterrows():
    prompt_dhcp = template_dhcp.rstrip() + ' If a user typed \'' + row['dhcp_hostname'] + '\', which is a technology vendor or a technology product name. What is the vendor the user typed? Consider that the user may misspell the vendor or provide invalid input. \nBob: '

    try:
        result = llm(prompt_dhcp)
    except:
        result = 'fail_to_process'

    df_llama2_dhcp.at[index, 'llama2_dhcp'] = result
    
    # Increment row counter
    row_counter += 1
    
    # Print message every 100 rows
    if row_counter % 100 == 0:
        print(f"Processed {row_counter} rows")

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Done! Elapsed Time: {elapsed_time} seconds")
```

```python
import time

start_time = time.time()

# Counter for printing messages
row_counter = 0

# Apply the function to each row
for index, row in df_llama2_netdisco.iterrows():
    prompt = template + row['netdisco_info'] + '\nBob: '
    
    try:
        result = llm(prompt)
    except:
        result = 'fail_to_process'

    df_llama2_netdisco.at[index, 'llama2_netdisco_vendor'] = result

    # Increment row counter
    row_counter += 1
    
    # Print message every 100 rows
    if row_counter % 100 == 0:
        print(f"Processed {row_counter} rows")

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Done! Elapsed Time: {elapsed_time} seconds")
```
